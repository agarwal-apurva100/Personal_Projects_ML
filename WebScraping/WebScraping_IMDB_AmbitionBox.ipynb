```python
# go to te webpage rightclick and then inspect the part u want to fetch from there we can get h2.
for i in soup.findAll('h2'):
    print(i.text.strip())      # strip will give 1 by 1
```

```python
for i in soup.findAll('p'):
    print(i.text.strip()) 
```

```python
# we will do the same for all 333 pages
company=soup.findAll('div',class_='company-content-wrapper')  # we have given class_ as class is a keyword in python we dont want that
```

```python
#<div class="company-content-wrapper">
```

```python
len(company)  # this is on one page
```

```python
company
```

```python
# We need all above feild
name=[]
rating=[]
review=[]
ctype=[]
headquarter=[]
how_old=[]
no_of_emp=[]

for i in company:
    name.append(i.find('h2').text.strip())   # when single info then find otherwise findAll
    rating.append(i.find('p',class_='rating').text.strip())
    review.append(i.find('a',class_='review-count').text.strip())
    ctype.append(i.findAll('p',class_='infoEntity')[0].text.strip()) 
    # as for all 4 ctype,headquarter,age and no is same name info entity so we did findall. [0]-- first data
    headquarter.append(i.findAll('p',class_='infoEntity')[1].text.strip()) 
    how_old.append(i.findAll('p',class_='infoEntity')[2].text.strip()) 
    no_of_emp.append(i.findAll('p',class_='infoEntity')[3].text.strip()) 
```

```python
# everything is coming in list format
# we will convery to df
df=pd.DataFrame({'Company_Name':name,
                'Comp_Rating':rating,
                'Comp_Review':review,
                'Comp_Type':ctype,
                'Comp_HeadQuarter':headquarter,
                'How_Old_Comp':how_old,
                'No_of_Emp_Working':no_of_emp,})
```
