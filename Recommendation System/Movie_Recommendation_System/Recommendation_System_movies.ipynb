```python
# we need to fix genre,cast,crew,keywords
# we need to import ast for handling this
import ast
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])    # we just needs name from genre
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(2)  # we can see we extracted it and in list format
```

```python
# rearrange the columns
movies=movies[['movie_id','title','genres','keywords','overview', 'cast', 'crew']]
movies.head(2)
```

```python
# now we will handle keywords
# keyword also name is only required , so we can apply the smae function
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# now we will handle cast and crew
# first cast-- we need three actors details, we just need the name
```

```python
movies['cast'][0]
```

```python
def convert3(text):
    L=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter<3:    # we can do 5 as well
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies.head(2)
```

```python
# similarly we need to handle crew
movies['crew'][0]
# we have job where the position shud be director
```

```python
def fetch_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
# this is our dataset after doing all things
```

```python
# we need ot convert to num
# there is one prob eg- actor is [Sam Worthington, and fdirector is [Sam Mendes]  so sam is there in both so it can impact the maodel
# so we will put the names as one name like (Sam Worthington) as one word
# what we will do we will remo ethe spaces from names 
```

```python
def collapse(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
    
```

```python
movies['cast']=movies['cast'].apply(collapse)
movies['crew']=movies['crew'].apply(collapse)
movies['genres']=movies['genres'].apply(collapse)
movies['keywords']=movies['keywords'].apply(collapse)
```

```python
from sklearn.feature_extraction.text import CountVectorizer  # BOW
cv=CountVectorizer(stop_words='english',max_features=5000,binary=True)       
# remove english stopwords
# we are just doing it on the top 5000 words
# we need only 0 and 1
```

```python
vector=cv.fit_transform(new['tags']).toarray()
```

```python
def recommend(movie):
    index=new[new['title']==movie].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])   # as we need values in reverse order
    for i in distances[1:6]:    # as we want top 5 movies
        print(new.iloc[i[0]].title)
# enumerate preserve the index
```

```python
recommend('Avengers: Age of Ultron')
```

```python
recommend('Harry Potter and the Half-Blood Prince')
```
